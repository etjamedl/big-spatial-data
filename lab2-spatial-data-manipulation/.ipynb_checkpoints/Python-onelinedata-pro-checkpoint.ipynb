#### Let check the different bands of the GSV image
A color image has three bands, red, green, and blue. Generally, green objects are brighter in the green band, red objects are brighter in the red band, and blue objects are brighter in the blue band. 

```python
I = im/255.0

red = I[:,:,0]
green = I[:,:,1]
blue = I[:,:,2]
```

Let check what different features look like in different bands.

```python
plt.subplots(1,3,figsize=(15,15))
plt.subplot(1,3,1)
plt.imshow(red)
plt.subplot(1,3,2)
plt.imshow(green)
plt.subplot(1,3,3)
plt.imshow(blue)
```

#### The street greenery are green in the GSV image. Based on this, we can define a rule to extract the greenery from Google Street View image.
In this example, we used the ExG = 2*Green - Blue - Green, to extract the greenery from the GSV image. 

```python
# calculate the difference between green band with other two bands
green_red_Diff = green - red
green_blue_Diff = green - blue

ExG = green_red_Diff + green_blue_Diff
plt.imshow(ExG)
```

It is clear that, greenery is brighter in the ExG image. Therefore, we can use a threshold to seperate the greenery and non-greenery

```python
greenery = ExG > 0.1
plt.imshow(greenery)
```

Calculate the percentage of greenery from the image

```python
greenPxlNum = len(np.where(greenery != 0)[0])
greenPercent = greenPxlNum/(400.0*400)*100
print ('The percentage of the greenery in the image is:', greenPercent)
```

#### Collect tweet with hashtag MIT
There are many many tweets in this case, we only access 10k tweet since after 2017

We can also collect only those geo-taged tweets. Unfortunately, you cannot access to the whole pool of tweet and only a small percentage of tweets are geo-tagged. So, you need be careful, the results you get may cannot fully representative to the reality. The spatial distribution pattern maybe biased, because of the samll samples. You need to very careful before making any conclusion based on your anlaysis on this dataset.

```python
results = []
geo_results = []

for tweet in tweepy.Cursor(api.search, q="#MIT",count=10000,
                           lang="en",
                           since="2017-01-01").items():
    
    results.append(tweet)
    
    # The time tag of the tweet
    time_tag = tweet.created_at
    
    # the location and coordinate of the user and tweet
    location = tweet.user.location
    coord = tweet.coordinates
    
    if coord is not None:
        print ('The location of the tweet is:', time_tag, location, coord)
        geo_results.append(tweet)
        
```

#### Convert the list into dataframe

```python
#Store tweets data in a dataframe
def tweets_df(results):
    id_list = [tweet.id for tweet in results]
    data_set = pd.DataFrame(id_list, columns = ["id"])
    
    data_set["text"] = [tweet.text for tweet in results]
    data_set["created_at"] = [tweet.created_at for tweet in results]
    data_set["retweet_count"] = [tweet.retweet_count for tweet in results]
    data_set["user_screen_name"] = [tweet.author.screen_name for tweet in results]
    data_set["user_followers_count"] = [tweet.author.followers_count for tweet in results]
    data_set["user_location"] = [tweet.author.location for tweet in results]
    data_set["Hashtags"] = [tweet.entities.get('hashtags') for tweet in results]
    data_set["coordinates"] = [tweet.coordinates for tweet in results]
    return data_set

data_set = tweets_df(results)
# data_set_geo = tweets_df(geo_results)
```
