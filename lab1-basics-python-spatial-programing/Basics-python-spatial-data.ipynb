### `for` loop

A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string). This is less like the for keyword in other programming languages, and works more like an iterator method as found in other object-orientated programming languages. With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc.

```python
cities = ['San Francisco', 'Los Angeles', 'New York', 'Atlanta']

for city in cities:
    print(city)
```

```python
# define a list and then loop each item
list_ex = [1, '1', 1.0, 'one', [1], [1,2,3]]
for i in list_ex:
    print(i)
```

### Open shapefile using `fiona`
However, the `geopandas` is not able to handle really big data and it doesn't give you more flexibility to handle the shapefile. Therefore, we can think about using lower-level (more difficult to program but give you more flexibility). Here is an example of using `fiona`, which is the base module of `geopandas`. 

```python
# loop all features in the shapefile using fiona
import fiona
from shapely.geometry import shape

datasource = fiona.open(shpfile)
crs = datasource.crs

# loop each feature
for feat in datasource:
#     print (feat['properties'])
    # get the properties of each feature
    props = feat['properties'] #here feat is a dictionary
    
    # Then you can get the attribute of the feature using dictionary key-pair
    perCapita = props['per_capita']
    tract = props['tract']
    
    # get the geometry of the feature
    geom = feat['geometry']
    geomObj = shape(geom)
    print('The per capita income of the tract', perCapita, tract)
    
```

### Open georeferenced Raster data using `rasterio`
`Geopandas` and `fiona` are good at handling the shapefiles. Using `rasterio` will help you to open and analyze raster data. The following script will read the raster data into a numpy array. Then you can manipulate and analyze the raster data.  

```python
import rasterio as rio

rasterfile = 'data/26928E233974N_dsm.tif'
raster = rio.open(rasterfile)

band = raster.read(1) #read the band 1, if you raster data is multi-bands, you can then get other bands
```

Check the dimension of the raster data

```python
band.shape
```

Then we can show the image of the raster

```python
plt.figure(figsize=(10, 10))
plt.imshow(band)
```
